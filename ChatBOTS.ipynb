```python
# Chatbot sederhana pakai dictionary
responses = {
    "halo": "Hai! Ada yang bisa saya bantu?",
    "apa kabar": "Saya baik, terima kasih! Anda?",
    "bye": "Sampai jumpa!"
}

def rule_based_chatbot(user_msg):
    key = user_msg.lower()
    return responses.get(key, "Maaf, saya belum paham.")

# Contoh pemakaian
print(rule_based_chatbot("halo"))
print(rule_based_chatbot("bagaimana"))
```

```python
# 1. Install dependencies
!pip install chatterbot chatterbot_corpus gradio

# 2. Import dan setup ChatBot
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer

chatbot = ChatBot(
    'SimpleBot',
    storage_adapter='chatterbot.storage.SQLStorageAdapter',
    database_uri='sqlite:///database.sqlite3'
)

trainer = ListTrainer(chatbot)

# 3. Data training yang lebih kaya
trainer.train([
    # Salam
    "Halo", "Halo! Ada yang bisa saya bantu?",
    "Hi", "Hai juga!",
    "Selamat pagi", "Selamat pagi! Semoga harimu menyenangkan.",
    "Selamat siang", "Selamat siang! Apa kabar?",

    # Tanya kabar
    "Apa kabar?", "Saya baik, terima kasih. Kamu sendiri bagaimana?",
    "Baik", "Senang mendengarnya!",
    "Tidak begitu baik", "Maaf mendengarnya. Semoga harimu segera membaik.",

    # Perkenalan
    "Siapa nama kamu?", "Nama saya SimpleBot, asisten virtualmu.",
    "Siapa kamu?", "Saya adalah chatbot yang dibuat untuk membantu Anda.",

    # Kemampuan
    "Apa yang bisa kamu lakukan?",
    "Saya bisa menjawab pertanyaan dasar, bercakap santai, dan mengingatkan Anda hal-hal sederhana.",

    # FAQ sederhana
    "Bagaimana cara memesan tiket kereta?",
    "Anda bisa memesan tiket kereta melalui KAI Access atau situs web kai.id.",

    # Small talk
    "Apa hobi kamu?", "Saya suka memproses bahasa manusia‚Äîalias ngobrol denganmu!",
    "Saya suka baca buku", "Bacaan apa yang sedang kamu nikmati sekarang?",

    # Perpisahan
    "Terima kasih", "Sama-sama, senang bisa membantu!",
    "Bye", "Sampai jumpa lagi!"
])

# 4. Buat fungsi wrapper untuk Gradio
def respond(message):
    response = chatbot.get_response(message)
    return str(response)

# 5. Bangun UI dengan Gradio
import gradio as gr

with gr.Blocks(theme=gr.themes.Default()) as demo:
    gr.Markdown("## ü§ñ SimpleBot Chat")
    gr.Markdown("> Chatbot ini dibangun dengan ChatterBot dan Gradio")

    chatbot_ui = gr.Chatbot()
    msg_in = gr.Textbox(
        placeholder="Ketik pesan di sini...",
        label="You"
    )
    send_btn = gr.Button("Kirim")

    def user_interaction(user_message, history):
        bot_reply = respond(user_message)
        history = history + [(user_message, bot_reply)]
        return "", history

    send_btn.click(
        fn=user_interaction,
        inputs=[msg_in, chatbot_ui],
        outputs=[msg_in, chatbot_ui]
    )
    # Meng-handle tekan Enter
    msg_in.submit(
        fn=user_interaction,
        inputs=[msg_in, chatbot_ui],
        outputs=[msg_in, chatbot_ui]
    )

demo.launch(share=True)
```
